```python
def cal_concat_mean_std (data_array: np.ndarray,train_size: float,val_size: float):
	num_time_steps = data_array.shape[0]
	num_train,num_val = (int (num_time_steps * train_size),int (num_time_steps * val_size),)
	# 计算标准化时启用
	train_array = data_array[:num_train]
	mean,std = train_array.mean (axis=0),train_array.std (axis=0)
	return mean,std
mean,std=cal_concat_mean_std(presurzer_data_concat,0.9,0.1)
```

```python
train_batch_dataset,val_batch_dataset,test_batch_dataset,train_datasets,train_label_sets,val_datasets,val_label_sets,test_datasets,test_label_sets,mean,std = lstm.get_data(True,presurzer_data,4,mean,std,3,True)
```

```python
import tensorflow as tf
from tqdm import tqdm
np.random.seed(32)  # set random seed for numpy
tf.random.set_seed(32)  # set random seed for tensorflow-cpu

save_data = []

for epoch in (range(2, 150, 2)):
	model = lstm.create_model (inputshape=(10,8),outputshape=4)
	check_point_fill = f'M:\work\project_program\\bishe\pyproject\RL_pid_control_critic_actor\强化学习+Lstm\稳压器水位压力控制\稳压器模型\\best_model_epoch{epoch}.hdf5'
	check_point_callback = tf.keras.callbacks.ModelCheckpoint (filepath=check_point_fill,moniter='loss',save_best_only=False,
															   save_weights_only=False)
	history = model.fit (train_batch_dataset,epochs=epoch,validation_data=val_batch_dataset,callbacks=[check_point_callback])

	state=np.array([train_datasets[400]])
	result=[state]

	for i in tqdm(range(400,1400)):
		n_state=model.predict(state)
		temp=n_state.tolist()
		temp[0].extend(list(train_datasets[i+1,9,4:]))
		temp=np.array(temp).reshape(1,1,8)
		new=np.concatenate((state,temp),axis=1)
		new=np.delete(new,0,1)
		result.append(new)
		state=new
	result=np.array(result)
	result=result.reshape(1001,10,8)
	#print(result.shape,train_datasets.shape)

	compare_data=np.concatenate((train_datasets[400:1401,-1,0:2].reshape(1001,2),result[:,-1,0:2].reshape(1001,2)),axis=1)
	dp.setfig(colume=2) # 参数预设置
	plt.subplot(1,2,1)
	plt.plot(compare_data[:,0], color='red', linestyle='solid', label='真实值') #此处label指的是Lengend或图例名称
	plt.plot(compare_data[:,2], color='green', linestyle='dashed', label='模型输出')
	plt.xlabel('时间')
	plt.ylabel('值')
	plt.legend()
	plt.subplot(1,2,2)
	plt.plot(compare_data[:,1], color='red', linestyle='solid', label='真实值') #此处label指的是Lengend或图例名称
	plt.plot(compare_data[:,3], color='green', linestyle='dashed', label='模型输出')
	plt.xlabel('时间')
	plt.ylabel('值')
	plt.legend()
	plt.tight_layout()
	plt.savefig(fname=f'模型迭代比较epoch_{str(epoch)}.svg',format='svg')
	plt.close()

	cal_data=pd.DataFrame(compare_data)
	corr=cal_data.corr()#.iloc[1][0]
	corr_1,corr_2=corr.iloc[0][2],corr.iloc[1][3]
	dis_1,dis_2=np.sum(abs(compare_data[:,0]-compare_data[:,2]))/1000,np.sum(abs(compare_data[:,1]-compare_data[:,3]))/1000
	save_data.append([epoch,corr_1,corr_2,dis_1,dis_2])
np.save(r"epoch与相关性和平均误差的关系.npy",save_data)
```

```python
def runavg(x, width):
#输入一原始时间序列和宽度，返回一个平滑以后的时间序列
    n = len(x)
    x = np.append(x, np.append(x, x))
    x_smooth = np.convolve(x, np.ones(width)/width, mode='same')
    xs = x_smooth[n:2*n-3]
    return xs

dp.setfig(colume=1)
plt.figure(figsize=(4,4))
plt.subplot(2,2,1)
plt.plot(aaa.reshape(1,501).T,lw=1,alpha=0.5)
plt.plot(runavg(aaa.reshape(1,501).T,10), color='r',lw=0.8,label='喷淋阀开度')
plt.legend()
plt.subplot(2,2,2)
plt.plot(bbb.reshape(1,501).T,lw=1,alpha=0.5)
plt.plot(runavg(bbb.reshape(1,501).T,10), color='g',lw=0.8,label='蒸汽排放阀开度')
plt.legend()
plt.subplot(2,2,3)
plt.plot(ccc.reshape(1,501).T,lw=1,alpha=0.5)
plt.plot(runavg(ccc.reshape(1,501).T,10), color='b',lw=0.8,label='上冲流量调节阀开度')
plt.legend()
plt.subplot(2,2,4)
plt.plot(ddd.reshape(1,501).T,lw=1,alpha=0.5)
plt.plot(runavg(ddd.reshape(1,501).T,10), color='y',lw=0.8,label='电加热器数量')
plt.legend()
plt.tight_layout()
plt.savefig('稳压器控制变量.svg',format='svg')
plt.show()



```
