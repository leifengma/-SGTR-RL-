```python
import numpy as np

speed=np.load(r"speed_rec.npy")
```

```python
speed.shape
```

```python

speed=speed.reshape(len(speed),1)
speed.shape
```

```python

speed=np.append(speed,[[-56]]*1000,axis=1)


np.savetxt("speed.csv", speed,delimiter=",")
```

```python
import matplotlib.pyplot as plt
plt.figure (figsize=(10,8))
plt.title ("temp_speed_error",fontsize=20)
plt.plot(speed-(-56),label="error")
plt.plot([0]*len(speed),'r--',label="zero")
plt.legend (fontsize=15)
plt.show()
#plt.savefig("降温速率误差.svg",dpi=300, bbox_inches='tight')
```

```python
import matplotlib.pyplot as plt
plt.figure (figsize=(10,8))
plt.title ("temp_speed_error",fontsize=20)
plt.plot(speed[100:300]-(-56),label="error")
plt.plot([0]*len(speed[100:300]),'r--',label="zero")
plt.legend (fontsize=15)
plt.savefig("fangda降温速率误差.svg",dpi=300, bbox_inches='tight')
```
